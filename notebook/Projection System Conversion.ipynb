```python
home_dir = Path.home()
pathloc = 'Dropbox (Personal)/Consulting/Tenokonda/data/sampleData/CSV Files/6_Well_Master.csv'
path = home_dir.joinpath(pathloc)

#read the csv files
df = pd.read_csv(path, sep=',')

#read excel file
#df = pd.read_excel(path,sheet_name=0)

df.head()
```

```python
epsgRef = 'epsg:2027'
Xcol = 'XSurf'
Ycol = 'YSurf'
```

```python
# for loop to enter each row and preform the pyprog XY to lat lon conversion.
for index, row in df.iterrows():
    # epsg coord system (from EPSG row)
    inProj = Proj(init=epsgRef)
    # espg coord system for lat long
    outProj = Proj(init='epsg:4326')
    # X and Y coords (from X and Y rows)
    x1,y1 = row[Xcol],row[Ycol]#output
    x2,y2 = transform(inProj,outProj,x1,y1)
    #print (x2,y2)
    # create and fill in lat and long columns
    df.loc[index,'longitude'] = x2
    df.loc[index,'latitude'] = y2
    #print (row['X'],(row['EPSG']))
```
